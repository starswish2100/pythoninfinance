```python
import pandas as pd
import numpy as np
import os
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
# Import all the daily csv file
import glob

path = r'../data' # use your path
all_files = glob.glob(path + "/*.xlsx")

df.empty

li = []

for filename in all_files:
    df = pd.read_excel(filename, sheet_name='Worksheet')
    if filename.endswith('.xlsx'):
        df['Date']=os.path.basename(filename)[:-5]
        #print(df['filename'])
        
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
print(frame)

```

```python
# Pick one paticular Term and remove the duplication entries
Term3MO = frame[frame['Term']=='3 MO'].drop_duplicates()
Term3MO.head(100)
```

```python
# Generate trend for the selected loan 

#fig, ax = plt.subplots(figsize=(15,7))
# use unstack()
#Term3MO.groupby(['Date','Term']).count()['Discount'].unstack().plot(ax=ax)
ax=Term3MO.plot(x='Date', y='Discount', kind = 'line')
mean=Term3MO['Discount'].mean()
standard=Term3MO['Discount'].std()
ax.axhline(y=mean, xmin =-1, xmax=1, color='g' , linestyle='--', lw=2)
ax.axhline(y=mean-standard, xmin =-1, xmax=1, color='r' , linestyle='--', lw=2)
ax.axhline(y=mean+standard, xmin =-1, xmax=1, color='r' , linestyle='--', lw=2)
```
